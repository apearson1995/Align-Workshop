# Fibonnaci Time

```python
# Factorial 
"""
0! = 1
1! = 1
2! = 2*1
3! = 3*2*1
"""
def factorial(num):
    i = 1
    mult = 1
    while i <= num:
        mult = mult * i
        i = i + 1
    return mult
 
```

# More Sophisticated Distance

See if you can understand why the below code works, it should make sense if you understand prop_diff.

```python
"""
Compute proportion of transitions from two strings of the of the same length. 

Transitions are  (A<->G, C<->T)

Args: 
    s1: string 1
    s2: string 2

Return:
    Error if the strings are not of same length
    Else, return a and b of different letters
"""
def prop_trs(s1,s2):
    _sum = 0
    i = 0
    while i < len(s1):
        if s1[i] == A:
            if s2[i] == G:
                _sum = _sum + 1
        elif s1[i] == G:
            if s2[i] == A:
                _sum = _sum + 1
        elif s1[i] == C:
            if s2[i] == T:
                _sum = _sum + 1
        elif s1[i] == T:
            if s2[i] == C:
                _sum = _sum + 1
    return _sum/len(s1)
```

```python
"""
Compute proportion of transversions from two strings of the same length. 

Transversions are (A<->C, A<->T, C<->G, G<->T).

Args: 
    s1: string 1
    s2: string 2

Return:
    Error if the strings are not of same length
    Else, return a and b of different letters
"""
def prop_trv(s1,s2):
    # Your Code Here
    _sum = 0
    i = 0
    while i < len(s1):
        if s1[i] == A:
            if s2[i] == C:
                _sum = _sum + 1
        elif s1[i] == C:
            if s2[i] == A:
                _sum = _sum + 1
        elif s1[i] == A:
            if s2[i] == T:
                _sum = _sum + 1
        elif s1[i] == T:
            if s2[i] == A:
                _sum = _sum + 1
        if s1[i] == C:
            if s2[i] == G:
                _sum = _sum + 1
        elif s1[i] == G:
            if s2[i] == C:
                _sum = _sum + 1
        elif s1[i] == G:
            if s2[i] == T:
                _sum = _sum + 1
        elif s1[i] == T:
            if s2[i] == G:
                _sum = _sum + 1
    return _sum/len(s1)
```
